# 4. Read toothpaste sales data of each month and show it using a scatter plot

image.png

```python
plt.figure(figsize=(10,7))

plt.scatter(df['month_number'],df['toothpaste'])

plt.title('Tooth paste Sale data',fontsize = 20)

plt.legend(['Tooth paste Sales data'],fontsize = 18)

plt.xlabel('Month Number',fontsize=16)

plt.ylabel('Number of units sold',fontsize = 16)

plt.xticks(fontsize = 16)

plt.yticks(fontsize = 16)

plt.xticks(df['month_number']) #show all X

plt.grid(linestyle='--')

plt.show()
```

# 5. Read face cream and face wash product sales data and compare them using the bar chart

image.png

```python
plt.figure(figsize=(10,7))

w = 0.4



bar1 = np.arange(len(df.month_number))

bar2 = [i+w for i in bar1]



plt.bar(bar1,df['facecream'],w,label='Face Cream sales data')

plt.bar(bar2,df['facewash'],w,label = 'Wash Cream sales data')

plt.xticks(bar1+w/2,df.month_number,fontsize=16)

plt.yticks(fontsize=16)

plt.xlabel('Month Numer',fontsize = 16)

plt.ylabel('Sales unit in number',fontsize = 16)

plt.legend(fontsize = 16,loc = "upper left")

plt.title('Facewash and facecream sales data',fontsize = 16)

plt.grid(linestyle = "--")
```

# 8. Calculate total sale data for each product and show it using a Pie chart

image.png

```python
df.sum()[1:7]
```

```python
# 8 
df.columns[1:7]
lst = []
for i in df.columns[1:7] :
    lst.append(sum(df[i]))

df2 = pd.DataFrame([
lst
], columns = df.columns[1:7])

y = lst
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.pie(y,autopct='%1.1f%%', labels=df2.columns, textprops={'fontsize': 20})
plt.legend(loc=4 ,prop={'size': 15})
plt.title('Sales Data', fontsize = 20)
plt.show()
```

# 9. Read Bathing soap and facewash of all months and display it using the Subplot

image.png

```python
# 9
x = df['month_number']
y = df['bathingsoap']
z = df['facewash']
fig, (ax1, ax2) = plt.subplots(2, sharex=True)
#fig.suptitle('Sales data of a Bathingsoap')
ax1.plot(x, y,'-ko',linewidth=3)
ax1.set_title('Sales data of a Bathingsoap')
ax1.label_outer()
ax2.plot(x, z, '-ro',linewidth=3)
ax2.set_title('Sales data of a Facewash')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(x)
plt.show()
```
