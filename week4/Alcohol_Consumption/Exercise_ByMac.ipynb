# GroupBy

### Introduction:

GroupBy can be summarized as Split-Apply-Combine.

Special thanks to: https://github.com/justmarkham for sharing the dataset and materials.

Check out this [Diagram](http://i.imgur.com/yjNkiwL.png)  
### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

### Step 3. Assign it to a variable called drinks.

```python
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
df.head()
```

### Step 4. Which continent drinks more beer on average?

```python
df.groupby('continent').agg({'beer_servings' : 'mean'})
```

```python
df.groupby('continent')['beer_servings'].mean()
```

### Step 5. For each continent print the statistics for wine consumption.

```python
df.groupby('continent').wine_servings.describe()
#type(df.groupby('continent').wine_servings.describe())
```

```python
df.groupby('continent').wine_servings.apply(lambda x : x.describe())
#type(df.groupby('continent').wine_servings.apply(lambda x : x.describe()))
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
df.groupby('continent').mean()
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
df.groupby('continent').median()
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
df.groupby('continent').agg({'spirit_servings':['mean', 'min', 'max']})
```

```python
df.groupby('continent').spirit_servings.agg(['mean', 'min', 'max'])
```
