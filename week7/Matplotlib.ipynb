```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 1. Line chart

```python
ts = pd.Series(np.random.randn(100), index=pd.date_range("1/1/2000", periods=100))
ts.plot();
```

```python
ts = ts.cumsum()
ts.plot();
```

```python
df = pd.DataFrame(np.random.randn(100, 4), index=ts.index, columns=list("ABCD"))

df = df.cumsum()

plt.figure();

df.plot();
```

## 2. Bar plot

```python
plt.figure();

df.iloc[0].plot(kind="bar");
```

Multiple bar plot

```python
df2 = pd.DataFrame(np.random.rand(10, 4), columns=["a", "b", "c", "d"])

df2.plot.bar();
```

```python
df2.plot.bar(stacked=True);
```

```python
df2.plot.barh(stacked=True);
```

## 3. Histogram

randn generates samples from the normal distribution, while numpy. random. rand from a uniform distribution (in the range [0,1))

```python
df4 = pd.DataFrame(
    {
        "a": np.random.randn(1000) + 1,
        "b": np.random.randn(1000),
        "c": np.random.randn(1000) - 1,
    },
    columns=["a", "b", "c"],
)


plt.figure();

df4.plot.hist(alpha=0.5);
```

## 5. Area plot

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=["a", "b", "c", "d"])

df.plot.area();
```

## 6. Scatter plot

```python
df = pd.DataFrame(np.random.rand(50, 4), columns=["a", "b", "c", "d"])

df["species"] = pd.Categorical(
    ["setosa"] * 20 + ["versicolor"] * 20 + ["virginica"] * 10
)

df.plot.scatter(x="a", y="b");
```

```python
ax = df.plot.scatter(x="a", y="b", color="DarkBlue", label="Group 1")

df.plot.scatter(x="c", y="d", color="DarkGreen", label="Group 2", ax=ax);
```

```python
df.plot.scatter(x="a", y="b", c="c", s=50);
```

```python
df.plot.scatter(x="a", y="b", c="species", cmap="viridis", s=50);
```

```python
df.plot.scatter(x="a", y="b", s=df["c"] * 200);
```

## 7. Pie chart

```python
series = pd.Series(3 * np.random.rand(4), index=["a", "b", "c", "d"], name="series")
series.plot.pie(figsize=(6, 6));
```

```python
df = pd.DataFrame(
    3 * np.random.rand(4, 2), index=["a", "b", "c", "d"], columns=["x", "y"]
)

df.plot.pie(subplots=True, figsize=(8, 4));
```

```python
series.plot.pie(
    labels=["AA", "BB", "CC", "DD"],
    colors=["r", "g", "b", "c"],
    autopct="%.1f",
    fontsize=20,
    figsize=(6, 6),
);
```

# Plotting tools
## 8. Scatter matrix plot

```python
from pandas.plotting import scatter_matrix

df = pd.DataFrame(np.random.randn(1000, 4), columns=["a", "b", "c", "d"])

scatter_matrix(df, alpha=0.2, figsize=(6, 6), diagonal="kde");
```

### Density plot
