```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7))
# the histogram of the data
ax.hist(x,50, density=1, facecolor='C0', alpha=0.75)

ax.set_xlabel('Length [cm]', fontsize=20, color='blue')
ax.set_ylabel('Probability')
ax.set_title('Aardvark lengths\n (not really)')
ax.text(75, .025, r'$\mu=115,\ \sigma=15$')
ax.axis([55, 175, 0, 0.03])
ax.grid(True);
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)  # make 4 random data sets
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))
ax1.plot(data1)
ax2.plot(data2, c='r')
```

```python
fig, ax = plt.subplots(1,2, figsize = (5,2.7))
ax[0].plot(data1)
ax[0].set_yscale('log')

ax[1].scatter(data2,data3)
ax[1].set_ylim(-5,5)
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = ax.plot(t, s, lw=2)

ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)  # make 4 random data sets
fig, ax = plt.subplots( figsize=(5, 2.7))
ax.plot(data1,'d',c='k' , label='d1')
ax.plot(data2, c='r', label='dd2')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.boxplot([data1,data2])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = ax.plot(t, s, lw=2)

ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_ylim(-2, 2);
```

```python
fig, axs = plt.subplots(1, 2, figsize=(5, 2.7))
xdata = np.arange(len(data1))  # make an ordinal for this
data = 10**data1
axs[0].plot(xdata, data)

axs[1].set_yscale('log')
axs[1].plot(xdata, data);
```

```python
fig, axs = plt.subplots(2, 1)
axs[0].plot(xdata, data1)
axs[0].set_title('Automatic ticks')

axs[1].plot(xdata, data1)
axs[1].set_xticks(np.arange(0, 100, 30), ['zero', '30', 'sixty', '90'])
axs[1].set_yticks([-1.5, 0, 1.5])  # note that we don't need to specify labels
axs[1].set_title('Manual ticks')
```
