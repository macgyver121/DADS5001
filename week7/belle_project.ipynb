```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('/content/drive/MyDrive/data_b.csv')
```

```python
df.head()
```

```python
df['category'].value_counts()
```

```python
df['category'] = df['category'].replace(['Black'], 'Black Twining')
df['category'] = df['category'].replace(['green'], 'Green Twining')
df['category'] = df['category'].replace(['energise'], 'Herbal (Energise)')
df['category'] = df['category'].replace(['gunpowder'], 'Gunpowder Green Tea ')
df['category'] = df['category'].replace(['recovery'], 'Herbal (Recovery)')
df['category'] = df['category'].replace(['hibiscus'], 'Hibiscus')
```

```python
df['category'].value_counts()
```

```python
df['Tea type'] = df['category']
```

```python
x = df.folin
y = df.teac
colors = {'Black':'black', 'green':'green', 'energise':'blue', 'gunpowder':'salmon', 'recovery':'grey', 'hibiscus':'red'}
fig, ax = plt.subplots(figsize = (9, 9))
ax.scatter(x, y, c=df['category'].map(colors))
m, b = np.polyfit(x, y, 1)
ax.plot(x, m*x + b)
ax.set_xlabel('Folin')
ax.set_ylabel('TEAC')
ax.set_title('TEAC - Folin')
plt.legend(df['category'].map(colors))
plt.show()
```

```python
sns.scatterplot(data=df, x = df['folin'], y = df['teac'], hue="category")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(figsize = (10,7))
sns.set_style("white")
sns.scatterplot(data=df, x = df['folin'], y = df['teac'], hue="Tea type", ax=ax, palette = 'Set1')
plt.setp(ax.get_legend().get_texts(), fontsize='13')
plt.setp(ax.get_legend().get_title(), fontsize='15')
ax2 = ax.twinx()
sns.regplot(x = df['folin'], y = df['teac'], data=df, ax=ax2, scatter=False, ci=None)
ax.set_xlabel('Phenol content', fontsize = 15)
ax.set_ylabel('Antioxidant capacity', fontsize = 15)
ax2.set(ylabel=None)
plt.savefig("photo.png", dpi = 400)
plt.show()
```
